````python
# https://matplotlib.org/stable/gallery/lines_bars_and_markers/simple_plot.html

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.svg")
plt.show()
````

#### Paths and Directories

````python
# Get current directory
directory = Path(os.getcwd())
directory
````

````python
# Combine directories with pathlib
directory_new = directory / 'subfolder'
directory_new
````

````python
# Create folder
def create_folder(folderpath):
    """
    Create a folder.

    Parameters
    ----------
    folderpath : Absolute path of the folder to be created. 
    """
    try: os.mkdir(folderpath)
    except OSError: print(f'Creation of the directory {folderpath} failed. Probably already present.')
    else: print(f'Successfully created the directory {folderpath}')
````

````python
create_folder(directory_new)
````

#### Finding and Handling Files

````python
# Find specific file types in current directory and subdirectories (recursively)
file_directory = Path(os.getcwd())

for path in file_directory.rglob('*.ipynb'):
    print(path)
````

````python
# Find specific file types in current directory
file_directory = Path(os.getcwd())

for path in file_directory.glob('*.ipynb'):
    print(path)
````

````python
# Get file name with extension (pathlib)
path.name
````

````python
# Get file name without extension (pathlib)
path.stem
````
